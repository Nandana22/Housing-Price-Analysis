### We are gonna continue from where we left off with the state level data

```python
import pandas as pd
import Quandl as qd
```

```python
main_df = pd.DataFrame()
state_tickers = []
us_states = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states')
for state in us_states[0][0][1:]:
    #state_tickers.append(''.join(['FMAC/HPI_',state]))
    query = "FMAC/HPI_"+str(state)
    #print(query)
    df = qd.get(query)
    print(df.head())
    # reset the index to make the index column into a regular Date column
    df = df.reset_index()
    # Rename the column name from value to the ticker
    df.rename(columns = {'Value':query}, inplace = True)
    
    # Initially, our main_df is empty. Assign df to it and then from the next iteration, merge the new df with our main_df
    if main_df.empty:
        main_df = df
    else:
        main_df = pd.merge(main_df, df, on='Date')
```
