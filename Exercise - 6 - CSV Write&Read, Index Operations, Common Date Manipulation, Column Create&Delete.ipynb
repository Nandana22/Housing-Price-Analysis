```python
import pandas as pd
```

```python
# Load the data from the csv file we created in the previous exercise.
df = pd.read_csv('FMAC_HPI_ALL_50_STATES.csv')
df.head()
```

```python
# Lets drop the unnamed column.
df = df.drop('Unnamed: 0', axis=1)
```

```python
# Lets create a new column called year that contains year info alone
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
```

```python
df.head()
```

```python
# Our month column contains only numerical representation of the months. But we need the actual names of the months
df['Month'] = df['Date'].dt.strftime('%b')
```

#### Oops. Looks like our date column is still not of the type date. Let's convert it into a date column.
df['Date'] = df['Date'].astype('datetime64[ns]')

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
# Lets look at the data type now
# df.dtypes
# We could use the head command as well to see only the first few columns and their dtypes.
df.dtypes
```

```python
# Exellent, we have our Date column all set up now.Let's run that command again.
df['Month'] = df['Date'].dt.strftime('%b')
```

```python
df.head()
```

#### That worked a charm again. Now, we could create a year-month column from this data.

```python
# The problem, with the above command is that the new column month will be of type object rather than String a we'd expect
# This is because the pandas dataframe stores the pointers to the strings and hence it is of type 'object'. 
# This might create problems at a later point of time when we need to concatenate strings, as we would  do
# when we try to create a "Period" column by combining the Year and Month data.

# So, let's change the dtype of Month column right away.
```

```python
df['Month'] = df.Month.apply(str)
```

```python
df.Month.dtypes
```

#### Now, lets try and create the new column by combining the Year and Month column.

```python
df['Period'] = df[['Year', 'Month']].apply(lambda x: ''.join(x), axis=1)
```

### There we go again, we ran into another problem. We need to convert the year column to be a string type.

```python
df.Year.dtype
```

```python
# It is of int type still.
```

```python
# Lets use the cat() to get the job done.
# df['Year-Month'] = df.Year.str.cat(df.Month)
df['Period'] = df.Year.astype(str).cat(df.Quarter.astype(str))
```

#### Here we are taking each row of the df[['Year', 'quarter']] and passing it as series to join, and join concatenates the elements in the series. 

## That didn't work either. Check the documentation for the cat() command.
http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.cat.html

```python
df['Period'] = df.Year.map(str) + "-" + df.Month
```

```python
df.head()
```

```python
# Lets make 'Period' our index column and remove the Year and Month columns.
df = df.drop('Year', axis=1)
```
